## Simplicial Complex = Simplex Tree

```python
import numpy as np
import pandas as pd 
import gudhi as gd 
from sklearn import manifold 
# from pylab import *
```

Insert() outputs true unless the simplex is already in the tree.

```python
st = gd.SimplexTree()
st.insert([0,1])
st.insert([1,2])
st.insert([3,1])
```

```python
st.insert([1,3])
```

```python
st_gen = st.get_filtration()
for simplex in st_gen:
    print(simplex)
```

```python
print(f"Is [2,3] in the simplex?\n {st.find([2,3])}")
```

```python
print(f"Dimension: {st.dimension()}")
print(f"Number of Vertices: {st.num_vertices()}")
print(f"Number of Simplices: {st.num_simplices()}")
```

```python
print("d-skeleton:")
d_skeleton = st.get_skeleton(1)
print(d_skeleton)
```

Filtered Simplicial Complex: the filtration value of a vertex must be $\leq$ the value of edges connected to it, because the vertex is always added "before" the edges.

```python
print("Increasing value of node 3")
st.assign_filtration([3], filtration = 0.8)
for simplex in st.get_filtration(): print(simplex)

print("\nMaking simplex filtered again increases edge value:")
st.make_filtration_non_decreasing()
for simplex in st.get_filtration(): print(simplex)
```

Inserting with filtration values

```python
st.insert([1,2,3], filtration = 0.1)
```

```python
print(f"Filtration Value of [1,2,3]: {st.filtration([1,2,3])}")
print(f"Dimension: {st.dimension()}")
print(f"Number of Vertices: {st.num_vertices()}")
print(f"Number of Simplices: {st.num_simplices()}")
```

```python
st.insert([0,1,2,3], filtration=0.2)
```

```python
print(f"Dimension: {st.dimension()}")
print(f"Number of Vertices: {st.num_vertices()}")
print(f"Number of Simplices: {st.num_simplices()}")
```
